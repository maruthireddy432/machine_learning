## Scaling data

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
std_scaler=StandardScaler()
mm_scaler=MinMaxScaler()
num_df=df.select_dtypes(include='int')
print("Scaled using standard scaler\n",std_scaler.fit_transform(num_df),'\n')
print('Scaled using min-max scaler\n',mm_scaler.fit_transform(num_df))
```

In standard scaler each value is normalized by subtracting it's mean and dividing by standard deviation. It normalize the data disturbution.

Where in min-max scaler it scales its feature to range of '0' to '1', Based on data disturbution we can which one want to use.

## Encoding

```python
cat_df=df.select_dtypes(include='object')
cat_df
```

```python
len(cat_df['income'].unique())
```

```python
oh=[]
la=[]
for i in cat_df:
    if (len(cat_df[i].unique()))<=5:
        oh.append(i)
    else:
        la.append(i)
print('columns less than 5 categories:',oh)
print('columns greater than 5 categories:',la)        
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import LabelEncoder
one_encoder=OneHotEncoder(sparse_output=False)
lab_encoder=LabelEncoder()
onehot_data=pd.DataFrame(one_encoder.fit_transform(df[oh]),columns=one_encoder.get_feature_names_out(oh))
onehot_data
```

The onehot encoder creates seperate column for every category.It mostly sutable for lesser number of categories.

```python
label_data=pd.DataFrame()
for col in la:
    label_data[col]=lab_encoder.fit_transform(df[col])
label_data    
```

The label encoder gives a label for every category when more of categories are there we use label encoding for those columns. 

Generating new features

As we have age and education number from them we can get work experience for them

```python
df['work_experience']=df['age']-df['education_num']
df
```

```python
# capital_gain,capital_loss are skewed so that we can apply log transformation to make them normal distributed
df['log_capital_gain'] = np.log1p(df['capital_gain'])
df['log_capital_loss'] = np.log1p(df['capital_loss'])
df
```

## Implementing Isolation Forest for Outlier Detection

```python
from sklearn.ensemble import IsolationForest

num_col = df[['age', 'capital_gain', 'capital_loss', 'hours_per_week']]

iso_forest = IsolationForest(contamination=0.01, random_state=42)  
df['outliers'] = iso_forest.fit_predict(num_col)


cleaned_df = df[df['outliers'] == 1]
cleaned_df
```

Outliers in the data can make the model to incorrect predictions and also cause it to overfit the models
