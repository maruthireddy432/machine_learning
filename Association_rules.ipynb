```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori,association_rules 
from mlxtend.preprocessing import TransactionEncoder
```

```python
df=pd.read_excel("Online retail.xlsx")
df
```

```python
df.columns=['items']
```

```python
df
```

# pre-processing

```python
items=df['items'].apply(lambda x:x.split(','))
items=items.tolist()
```

```python
te=TransactionEncoder()
tedata=te.fit_transform(items)
df_transformed=pd.DataFrame(tedata,columns=te.columns_,dtype='int')
df_transformed
```

```python
frequent_itemsets=apriori(df_transformed,min_support=0.01,use_colnames=True)
frequent_itemsets
```

```python
rules=association_rules(frequent_itemsets,metric='lift',min_threshold=1)
rules.sort_values('lift',ascending=False).head(10)
```
