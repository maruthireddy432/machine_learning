## Encoding variables

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
for col in df.columns:
    if df[col].dtype=='object':
        df[col]=encoder.fit_transform(df[col])
```

```python
df.corr()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(13, 5))
sns.histplot(df['stalk_height'], kde=True, ax=axes[0]).set_title('Stalk Height Distribution')
sns.histplot(df['cap_diameter'], kde=True, ax=axes[1]).set_title('Cap Diameter Distribution')
plt.tight_layout()
```

## Train Test split

```python
from sklearn.model_selection import train_test_split
x=df.drop(columns='class')
y=df['class']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
```

```python
x_train.shape,y_train.shape
```

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.model_selection import cross_val_score,GridSearchCV
```

## Grid searchCV

```python
clf = SVC()
param_grid = [{'kernel': ['rbf'], 'gamma': [50, 5, 10, 0.5], 'C': [15, 14, 13, 12, 11, 10, 0.1, 0.01]}]

gsv = GridSearchCV(clf, param_grid=param_grid, cv=10)
gsv.fit(x_train, y_train)
```

## Best score and parameter values

```python
print(gsv.best_params_)
print(gsv.best_score_)
```

```python
clf=SVC(gamma=50,C=15)
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
acc=accuracy_score(y_test,y_pred)*100
print("Accuracy:",acc)
confusion_matrix(y_test,y_pred)
```

## Checking for different kernels

```python
clf=SVC(kernel='poly')
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
acc=accuracy_score(y_test,y_pred)*100
print("Accuracy:",acc)
confusion_matrix(y_test,y_pred)
```

```python
clf=SVC(kernel='sigmoid')
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
acc=accuracy_score(y_test,y_pred)*100
print("Accuracy:",acc)
confusion_matrix(y_test,y_pred)
```

```python
clf=SVC(kernel='rbf')
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
acc=accuracy_score(y_test,y_pred)*100
print("Accuracy:",acc)
confusion_matrix(y_test,y_pred)
```
